```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

from figure_formatting import figure_formatting as ff

ff.set_formatting()
```

```python
def conc_A(t, A0, k):
    return A0 * np.exp(-k*t)

def conc_B(t, A0, k):
    return A0 * (1 - np.exp(-k*t))
```

```python
t = np.linspace(0, 100)
k = 3e-2
A0 = 1

# Create figure and axis objects explicitly
fig, ax = plt.subplots()

# Plot the data
ax.plot(t, conc_A(t, A0, k), color='#4ca98f')
ax.plot(t, conc_B(t, A0, k), color='#EC6A06')

# Set axis limits
ax.set_xlim(0, 100)
ax.set_ylim(0, 1)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Remove x-axis ticks and labels
ax.set_xticks([])

# Label axes (using LaTeX notation)
ax.set_xlabel('time')
ax.set_ylabel('concentration / $[\mathrm{A}]_0$')

plt.savefig('first_order_example.pdf')

plt.show()
```
